### Funktion, die die http-Suppe einer url parsed und als soup ausgibt

```python
# Abfrage als http-Request (anstatt als Klick-Link mit selenium)

delay = 0.12 # or # Verzögerung festlegen
# treat this as pseudo-code as there's no link here for me to
# get back to your original article :( In particular, I don't
# recall what the exact value your passed as 2nd param to 
# BeautifySoup() I've only ever used 'lxml', but that would be
# an unnecessary dependency to add for your needs
def get_soup(url):
    sleep(delay)
    req = requests.get(url)
    global soup
    soup = BeautifulSoup(req.text, 'html.parser') 
    return soup
```

### Schleife, die alle einzelnen Job-Links in Sammeliste urls ausgibt.

```python
urls = [] # Sammelliste für alle einzelnen Job-Links
searches = [] # Sammelliste, die später die Zuordnung von Search-Link zu den Daten ermöglicht
num_jobs_list = []
job_count = []
# hier noch eine Pause in die Schleife bauen, die immer nur 10-50 Links abklappert, dann sleep
start = datetime.datetime.now()
print(start)
for link in search_sample: # hier ggfs wieder auf komplette search_list ausweiten
    try:
        get_soup(link) # beinhaltet sleep!
        # num_jobs = int((str(soup.find('b').get_text())).replace('Treffer für','') # ersetzt weil Ergebnisseite für Erweiterte Suche anders aussieht als für normale Suche
        num_jobs = int(soup.find('span', {'class': "at-facet-header-total-results"}).get_text())
        num_jobs_list.append(num_jobs)
        # gibt Anzahl der Stellen wieder, die auf der ersten Ergebnisseite steht und speichert diese im Objekt num_jobs und in der Liste num_jobs_list
        num_pages = math.ceil(num_jobs/(100)) ## 100 jobs pro Ergebnisseite --> berechnet Anzahl der Ergebnisseiten
        page_counter = [] # Liste, die die "Startzahlen" für die einzelnen Ergebnisseiten bereitstellt
        for i in range(0, num_pages+1, 1): 
            sleep(0.1)
            count = i*100
            page_counter.append(count)
        # Schleife, die durch alle Ergebnisseiten geht und dort alle Links einholt
        for i in page_counter:
            base_url = str(link) + "&of=" + str(i) # Abfrage iteriert durch of=0, 100, 200 usw durch die Ergebnisseiten
            try:
                soup = get_soup(base_url) # holt die Suppe der jeweiligen Ergebnisseite (sleep / delay = 1 Sekunde )
                articles = soup.findAll('article')[:num_jobs]
                for article in articles:
                    href = article.find('a').find_next('a').get('href') # zweiter Link immer der Title-Link
                    url = str("https://www.stepstone.de" + href) # auch dieser Scrape sieht in der Erweiterten Suche anders aus. 
                    urls.append(url) # hängt die gezogenen URLs jeder Seite an die Lists urls an
                    searches.append(link) #hängt den ursprünglichen search_link an die Liste searches an (muss für jeden Joblink gemacht werden, deshalb in der Funktion und nicht in der großen Schleife) # zupft alle Links von der jeweiligen Ergebnissseite und haengt sie an die Liste urls an
                    job_count.append(num_jobs) # anzahl der Jobergebnisse wird für jeden Link angehängt
            except:
                continue
    except:
            continue
end = datetime.datetime.now()
print(end)
print("Suchdauer: " + str(end - start))
```

### Funktion definieren, die Daten der einzelnen Jobs zieht.
* noch weitere vorstruktrierte Datenfelder generieren!
* Funktion zieht soup jeder einzelnen url und erstellt dann einzelne Objekte, fügt sie der globalen Listen hinzu
* Schöne Kurzanleitung: https://www.youtube.com/watch?v=v5DDW5dyfyc

```python
def get_posting(url):
    job_soup = get_soup(url) # beinhaltet sleep! 0.1 Sekunden (5 Minuten pro 1000 soups)
    # Suchlink des Jobs:
    job_url = url
    # Unternehmen:
    company = str(job_soup.find(name='a', attrs={'class': "at-listing-nav-company-name-link"}).get('title'))
    # company_list.append(company)
    company_link = job_soup.find('a', attrs={'at-listing-nav-company-name-link'}).get('href')
    # Orte:
    location = job_soup.find('li', attrs={'class': "listing-list at-listing__list-icons_location"}).get_text(strip=True)
    # location_list.append(str(location))
    # Anstellungsart (Festanstellung,...?)
    contract_type = job_soup.find('li', attrs={'class': "listing-list at-listing__list-icons_contract-type"}).get_text(strip=True)
    # contract_type_list.append(str(typ))
    # Arbeitszeit (VZ/TZ)
    wrk_time = job_soup.find('li', attrs={'class': "listing-list at-listing__list-icons_work-type"}).get_text(strip=True)
    # wrk_time_list.append(str(wrk_time))
    # Job-Anzeigen-Titel:
    title = job_soup.find(name='h1', attrs={'class': "listing__job-title js-edit-title at-listing-nav-listing-title"}).get_text(strip=True)
    # title_list.append(str(title))
    # Datum:
    date = job_soup.find('span', attrs={'class': "date-time-ago", "data-date": True})["data-date"][0:10]
    # date_list.append(str(date))
    # Detaillierte Jobbeschreibung (Kleintext-Absätze):                   
    sections = []
    for section in job_soup.find_all(name='section', attrs={'offer__section card card_ld branding-colors__secondary'}):
        txt = section.get_text(" ", strip=True)
        sections.append(txt) # diese 
    records.append((title, company, company_link, location, contract_type, wrk_time, date, job_url, sections))
    return('')
```

### Schleife, die alle Jobdetails / Postings aus den Links der Liste urls_unique zieht
* ACHTUNG: Schleife dauert sehr lange!!!
* evtl delay (sleep(delay)) einbauen, damit Server nicht blockiert!

```python
records = []
# hier noch eine Stopuhr einbauen
start = datetime.datetime.now()
print(start)
for i, url in enumerate(urls_sample): # enumerate iteriert über die Anzahl der Einträge in der Liste urls_unique später 
    try:
        get_posting(url)
        sleep(0.2) # bedeutet etwa 3,3 Minuten Verzögerung bei 1000 Links
        
    except: 
        continue
end = datetime.datetime.now()
print(end)
print("Laufzeit: " + str(end - start))
```
